## 2.1.3 广播机制
在某些情况下，即使形状不同，我们仍然可以通过调用 广播机制来执行按元素操作。 这种机制的工作方式如下：

1.通过适当复制元素来扩展一个或两个数组，以便在转换之后，两个张量具有相同的形状；

2.对生成的数组执行按元素操作。

```python
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a, b
```

```python
# 矩阵a将复制列， 矩阵b将复制行，然后再按元素相加
a + b 
```

## 2.1.5 节省内存

```python
before = id(Y)
Y = Y + X
id(Y) == before
```

```python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
```

```python
before = id(X)
X += Y
id(X) == before
```
