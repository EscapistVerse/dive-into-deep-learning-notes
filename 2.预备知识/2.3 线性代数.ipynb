## 2.3.10 范数

范数用于衡量向量的“大小”或“长度”。

---

## 范数的性质

### 1. 缩放性质

$$
f(\alpha x)=|\alpha|f(x)
$$

向量放大多少倍，范数也放大多少倍。

---

### 2. 三角不等式

$$
f(x+y)\le f(x)+f(y)
$$

直线距离不会大于绕路距离。

---

### 3. 非负性

$$
f(x)\ge0
$$

范数不可能为负数。

---

### 4. 零向量性质

$$
f(x)=0 \iff x=0
$$

只有零向量的范数为 0。

---

## L2 范数（欧几里得范数）

对于向量：

$$
x=[x_1,x_2,\dots,x_n]
$$

L2 范数定义为：

$$
||x||_2=
\sqrt{
\sum_{i=1}^{n}x_i^2
}
$$

即：

> 平方 → 求和 → 开根号

通常会省略下标 2：

$$
||x|| = ||x||_2
$$

```python
u = torch.tensor([3.0, -4.0])
torch.norm(u)
```
