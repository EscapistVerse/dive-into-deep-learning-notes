## 2.4.1 导数和微积分

书上代码由于环境版本兼容问题，于是这里换了一种画图方式

```python
import numpy as np

def f(x):
    return 3 * x ** 2 - 4 * x

x = np.arange(0, 3, 0.1)
y = f(x)
tangent_line = 2 * x - 3

from matplotlib.figure import Figure
import io
from PIL import Image
import IPython.display as display

fig = Figure(figsize=(6, 4.5))
ax = fig.add_subplot()

ax.plot(x, y, label='f(x)')
ax.plot(x, tangent_line, linestyle='--', label='Tangent line (x=1)')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.6)

buf = io.BytesIO()
fig.savefig(buf, format='png', bbox_inches='tight')
buf.seek(0)

img = Image.open(buf)
display.display(img)
```
